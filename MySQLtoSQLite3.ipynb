## DATA MIGRATION FROM MySQL TO SQLITE3

```python
# Import Libarary
import sqlite3
import mysql.connector as mysqlConnection
from pandas.io import sql
```

```python
# Establish Connection
sqliteConn = sqlite3.connect("testdb.db")

# MySQL Connection
mysqlConn = mysqlConnection.Connect(host="localhost", user="root", passwd="", database="beneficiary_db")
```

```python
# Declare Cursor
sqliteCursor = sqliteConn.cursor()
```

```python
# Excute MySQL Query, Fetch from MySQL
df = sql.read_sql("SELECT * FROM beneficiaries LIMIT 5", mysqlConn)
df
```

```python
# Load to SQLite Database
df.to_sql(name="testtbl", con=sqliteConn)
```

```python
# Fetch Single Column from SQLite Database
fetchedCursor = sqliteCursor.execute("SELECT * FROM testtbl;")
for i in fetchedCursor:
    print(i[15])
```

```python
# Fetch from SQLite Database
df = sql.read_sql("SELECT * FROM testtbl", sqliteConn)
df
```
